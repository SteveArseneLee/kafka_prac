```python
url = 'http://openapi.seoul.go.kr:8088/52576565426c636c3237564a424748/xml/VwsmTrdarSelngQq/'
date = []
name = []
# excerpt = []

for i in range(0,10):
    url_code = url + str(i*10+1) + '/' + str(i*10+10) + '/' + str('2020')
    html = urlopen(url_code)
    bs = BeautifulSoup(html, 'html.parser')
    svc_induty_cd_nm = bs.find_all('svc_induty_cd_nm')
    thsmon_selng_amt = bs.find_all('thsmon_selng_amt')
    print(url_code)
    
    for j in range(0,10):
        date.append(svc_induty_cd_nm[i].text)
        name.append(thsmon_selng_amt[i].text)
```

```python
import pandas as pd
DF = pd.DataFrame({'date':date, 'name':name})
DF
```
